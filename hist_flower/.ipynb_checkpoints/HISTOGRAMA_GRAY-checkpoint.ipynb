```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def hist_(path,title_):

    img = cv.imread(path)
    img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    hist = cv.calcHist([img_gray],[0],None,[256],[0,256])

    # compute and plot the image histograms
    # [max value to show] [number of bins] for cv.calHist

    plt.plot(hist)
        
    plt.title(title_)
    plt.show()
```

```python
hist_("img1_tp.png","DEGRADADO")
```

```python
hist_("img2_tp.png"," FLOR")
```

# Ambas imagenes tienen la misma distribución de valores de pixeles mas no en terminos de ubicación del mismo, incluso una imagen del mismo alto y ancho diferente a estas dos, podría presentar el mismo histograma.
